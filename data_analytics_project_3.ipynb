```python
column_name=input("enter the column name foe histogram")
plt.figure(figsize=(10,6))
sns.histplot(df['SepalLengthCm'],bins=30, kde=False, color='blue')
plt.xlabel(column_name)
plt.ylabel('SepalLengthCm')
plt.title(f'Distribution of {column_name}')
plt.grid(True)
plt.show()
```
